# 시리즈 활용

```python
import numpy as np
import pandas as pd

s = ['희빈', '수영', '현수', '지호', '지민']
p = pd.Series([99, 84, 92, 65, 78], index=s)
p
```

```python
p.name = '수학 성적'
p.index.name = '이름'
p
```

```python
p.sort_index()
```

```python
p.sort_index(ascending=False)
```

```python
p.sort_values()
```

```python
p.sort_values(ascending=False)
```

# 시리즈 기본 통계 함수

```python
print(p.max())
print(p.min())
print(p.mean())
print(p.median())
print(p.count())
print(p.size)
```

```python
p.describe()
```

```python
s = pd.Series([3, 1, 1, 2, None, np.nan, 4, 6, 6, np.nan])
s
```

```python
s.values
```

```python
s.count()
```

```python
s.value_counts()
```

```python
s.value_counts(dropna=False)
```

```python
s.value_counts(dropna=False, sort=False)
```

```python
s.value_counts(bins=4)
```

```python
s
```

```python
s.value_counts(bins=4, sort=False) # 1.0부터 6.0까지 4개의 구간을 나눔
```

```python
s.value_counts(bins=[0.5, 3.5, 6.5], sort=False)
```
